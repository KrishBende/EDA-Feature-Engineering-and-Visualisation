```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("FlightPredictionDataset/Data_Train.xlsx")
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop null entries since there is only 1.
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Converting categorical data to numerical data
#1 Airline
df['Airline'].unique()
pd.get_dummies(df['Airline'],dtype=int)
df=df.join(pd.get_dummies(df['Airline'],dtype=int))
df.drop(['Airline'],inplace=True,axis=1)
```

```python
#2 Date of Journey

df['Date']=df['Date_of_Journey'].str.split("/").str[0]
df['Month']=df['Date_of_Journey'].str.split("/").str[1]
df['Year']=df['Date_of_Journey'].str.split("/").str[2]

df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)

df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df['Source'].unique()
```

```python
#3 Source
pd.get_dummies(df['Source'],dtype=int)
df=df.join(pd.get_dummies(df['Source'],dtype=int, prefix="Source", prefix_sep="-"))
df.drop(['Source'],inplace=True,axis=1)

#4 Destination
df=df.join(pd.get_dummies(df['Destination'],dtype=int, prefix="Destination", prefix_sep="-"))
df.drop(['Destination'],inplace=True,axis=1)
```

```python
#5 Dep_Time
df['Departure_Hour']=df['Dep_Time'].str.split(":").str[0]
df['Departure_Min']=df['Dep_Time'].str.split(":").str[1]
df.drop(['Dep_Time'],inplace=True,axis=1)
```

```python
df
```

```python
#6 Arrival_Time
df['Arrival_Hour']=df['Arrival_Time'].str.split(":").str[0]
df['Arrival_Min']=df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[0]
df.drop(['Arrival_Time'],inplace=True,axis=1)
```

```python
#We can drop Duration, since it can be derived from Departure and Arrival Time
#We can also drop Route since we have the number of stops
df.drop(['Route','Duration'],axis=1,inplace=True)
```

```python
#7 Total Stops
df['Total_Stops'].unique()
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
#8 Additional Info

df['Additional_Info'].unique()
pd.get_dummies(df['Additional_Info'],dtype=int)
df=df.join(pd.get_dummies(df['Additional_Info'],dtype=int))
df.drop(['No Info','No info'],axis=1,inplace=True)
df.drop(['Additional_Info'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Departure_Hour']=df['Departure_Hour'].astype(int)
df['Departure_Min']=df['Departure_Min'].astype(int)
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Min']=df['Arrival_Min'].astype(int)
```

```python
#Splitting and Feature Scaling-
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Price'],axis=1), df['Price'], test_size=0.33, random_state=42)

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)

#Training models on the data-
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression().fit(X_train,y_train)
y_pred_linear=linear_model.predict(X_test)

from sklearn.tree import DecisionTreeRegressor
dtree_model=DecisionTreeRegressor().fit(X_train,y_train)
y_pred_dtree=dtree_model.predict(X_test)

#Accuracy
from sklearn.metrics import mean_squared_error
print("Linear Regression Model MSE- ",mean_squared_error(y_test,y_pred_linear))
print("Linear Regression Model MSE- ",mean_squared_error(y_test,y_pred_dtree))
```
